```python
%matplotlib inline
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import scipy.stats
import seaborn as sns
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import RANSACRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LogisticRegression
```

```python
iris = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data", header=None)
iris.columns = ["sep_l", "sep_w", "pet_l", "pet_w", "class"]
iris.head()
```

```python
iris_model = LogisticRegression(C = 1000)
```

```python
sns.set(style="ticks", color_codes=True)
sns.pairplot(iris, hue="class", palette="husl")
```

```python
test_indices = np.random.randint(0, len(iris), 5)
train_indices = np.delete(np.arange(len(iris)), test_indices)
iris_train = iris.ix[train_indices]
iris_test = iris.ix[test_indices]
# TODO: Split the labels from the other attributes
iris_model.fit(iris_train.drop("class", axis = 1), iris_train["class"])
```

```python
print(iris_model.predict(iris_test.drop("class", axis = 1)))
print(iris_test["class"])
```

We can have the probability for each of the classes

```python
print(iris_model.predict_proba(iris_test.drop("class", axis = 1)))
```
