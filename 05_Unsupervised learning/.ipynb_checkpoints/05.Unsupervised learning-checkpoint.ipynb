```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.datasets import make_blobs, make_moons, make_circles, make_regression, make_s_curve, make_swiss_roll
```

### 1. k-Means Clustering

```python
# clusters has the role of the labels and are used to check how good is working (algorithm)
# cluster_std - gives the dispersion (standard deviation), small value makes clusters more closed
attributes, clusters = make_blobs(cluster_std = 0.8, centers = 3)
```

```python
# attributes is array of arrays
attributes[:5]
```

```python
# it is a numbers representing each number from the attributes array
clusters
```

```python
plt.scatter(attributes[:, 0], attributes[:, 1], c = clusters)
```

#### Make moons

without noise parameter

```python
attributes, clusters = make_moons()
plt.scatter(attributes[:, 0], attributes[:, 1], c = clusters)
```

```python
# we have to put noise lowest from the signal (around 1)
attributes, clusters = make_moons(noise=0.08)
plt.scatter(attributes[:, 0], attributes[:, 1], c = clusters)
```

#### Make circles

without noise

```python
attributes, clusters = make_circles()
plt.axes().set_aspect('equal', 'datalim')
plt.scatter(attributes[:, 0], attributes[:, 1], c = clusters)
```

```python
attributes, clusters = make_circles(noise=0.08)
plt.scatter(attributes[:, 0], attributes[:, 1], c = clusters)
```

#### Other options - make_regression, make_s_curve, make_swiss_roll

```python
# multilevel regrerssion
attributes, clusters = make_regression()
plt.scatter(attributes[:, 0], attributes[:, 1], c = clusters)
```

```python
attributes, clusters = make_s_curve()
plt.scatter(attributes[:, 0], attributes[:, 2], c = clusters)
```

```python
attributes, clusters = make_swiss_roll()
# change - attributes[:, 2]
plt.scatter(attributes[:, 0], attributes[:, 2], c = clusters)
```
