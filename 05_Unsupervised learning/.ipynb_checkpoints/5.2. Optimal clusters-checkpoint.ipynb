### Finding optimal number of clusters - Elbow method

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from sklearn.datasets import make_blobs, make_moons #, make_circles
from sklearn.cluster import KMeans, AgglomerativeClustering
```

```python
attributes, clusters = make_blobs(centers=10, cluster_std=3.5)
# plt.axes().set_aspect('equal', 'datalim')
# plt.scatter(attributes[:, 0], attributes[:, 1], c = clusters)
# plt.show()

elbow_range  = 31

inertias = [] 
for i in range(1, elbow_range):
    km = KMeans(n_clusters = i)
    km.fit(attributes)
    inertias.append(km.inertia_)
plt.plot(range(1,elbow_range), inertias, marker = "o")
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
plt.show()
```

```python
attributes, clusters = make_blobs(cluster_std=2)



aggl = AgglomerativeClustering(n_clusters = 3, linkage = "complete")
aggl_predicted = aggl.fit_predict(attributes)
aggl_predicted

plt.scatter(attributes[:, 0], attributes[:, 1], c = clusters)
plt.show()
plt.scatter(attributes[:, 0], attributes[:, 1], c = aggl_predicted)
plt.show()
```

```python
attributes_moons, clusters_moons = make_moons(noise=0.05)

aggl_moons = AgglomerativeClustering(2, linkage = "ward")
aggl_predicted = aggl_moons.fit_predict(attributes_moons)
aggl_predicted

plt.scatter(attributes_moons[:, 0], attributes_moons[:, 1], c = clusters_moons)
plt.show()
plt.scatter(attributes_moons[:, 0], attributes_moons[:, 1], c = aggl_predicted)
plt.show()
```
